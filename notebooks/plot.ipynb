```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['Data Sync', 'Data Preprocessing', 'Phenotyping/Image(in secs)','Total phenotyping','Total']
time = [5.342, 17.953,1.01, 86.35,5.342+17.953+86.35]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x, time, width,label="Time")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Time elapsed in minutes')
ax.set_title('Processing time @ Intel(R) Xeon(R) CPU E5-2686 v4 2.30GHz and Telsa V100')

ax.set_xticks(x,labels)
ax.legend(loc="upper left")

ax.bar_label(rects1, padding=3)

fig.tight_layout()
fig.set_size_inches(8, 6)
plt.show()
fig.savefig("timing.png")
```
